# Andamento RMSE

Questo notebook visualizza, **solo per tutti i modelli**, l'andamento della metrica di test (RMSE medio) al variare del numero di epoche locali (1, 5, 10) per ciascun algoritmo federato: FedAvg_RUL, SCAFFOLD, FedProx_RUL, FedDyn_RUL, FedCross_RUL.

#### TASK A

```python
import matplotlib.pyplot as plt

local_epochs = [1, 5, 10]
colors = ["#6F2DBD", "#F29F05", "#0476D9", "#D92567", "#05A63D"]
colors = ["#F29F05", "#0476D9", "#D92567", "#05A63D", "#6F2DBD",]


# Valori di summary.global/test_rmse_mean (local_epochs = [1, 5, 10])
rmse_data = {
    "FedAvg":    [21.339386, 17.946732, 17.651504],
    "SCAFFOLD":  [24.699064, 22.317550, 20.767876],
    #"FedProx":   [21.329009, 17.953124, 17.643863],
    "FedDyn":    [28.567070, 28.306695, 27.536751],
    "FedCross":  [25.531702, 24.267412, 22.956018],
}

# Valori di summary.global/test_rmse_std (local_epochs = [1, 5, 10])
rmse_std = {
    "FedAvg":    [1.327673, 1.398346, 1.633875],
    "SCAFFOLD":  [7.614953, 9.292168, 9.372915],
    #"FedProx":   [1.324102, 1.446084, 1.612457],
    "FedDyn":    [6.922609, 6.678474, 6.525036],
    "FedCross":  [7.315163, 8.615863, 8.706822],
}
# 

# Valori di summary.global/test_rmse_mean (local_epochs = [1, 5, 10])
rmse_data = {
    "FedAvg":    [25.38664412342805, 19.114124237900263, 17.29128977991618],
    "SCAFFOLD":  [25.327895296995965, 19.441798195488687, 16.987894512009998],
    "FedDyn":    [35.57163202188154, 36.31832706845037, 36.14772657433513],
    "FedCross":  [25.752879531967565, 20.36620038919861, 19.092383982409192],
}

# Valori di summary.global/test_rmse_std (local_epochs = [1, 5, 10])
rmse_std = {
    "FedAvg":    [7.750007369476957, 2.1929932871207223, 1.6589928084205932],
    "SCAFFOLD":  [7.719003126703915, 2.5165873923352637, 1.846512346728873],
    "FedDyn":    [7.755095500148034, 7.701800068123281, 7.867876422282006],
    "FedCross":  [7.570404429236991, 2.6953718785929115, 2.5037754776539276],
}
```

```python
plt.figure(figsize=(8, 5))

for algo, values in rmse_data.items():
    std_values = rmse_std.get(algo)
    if std_values is not None:
        plt.errorbar(local_epochs, values, yerr=std_values, marker="o", capsize=4, label=algo)
    else:
        plt.plot(local_epochs, values, marker="o", label=algo)



plt.xlabel("Local epochs")
plt.ylabel("Test RMSE (mean ± std)")
plt.title("LSTM_RUL - Test RMSE vs Local Epochs (Task A)")
plt.xticks(local_epochs)
plt.grid(True, linestyle="--", alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Grafico a barre (Task A): mean RMSE con deviazione standard (error bar classiche)
local_epochs = [1, 5, 10]
x = np.arange(len(local_epochs))
algos = list(rmse_data.keys())
n_algos = len(algos)
bar_width = 0.74 / n_algos


fig, ax = plt.subplots(figsize=(6, 4))

for i, algo in enumerate(algos):
    means = rmse_data[algo]
    stds = rmse_std.get(algo, [0] * len(local_epochs))
    positions = x + (i - (n_algos - 1) / 2) * bar_width

    ax.bar(
        positions,
        means,
        width=bar_width,
        yerr=stds,
        capsize=4,
        alpha=0.85,
        label=algo,
        color=colors[i % len(colors)]
    )

ax.set_xlabel("Local epochs", fontsize=16)
ax.set_ylabel("RMSE", fontsize=16)
#ax.set_title("Task A")
ax.set_xticks(x)
ax.set_xticklabels(local_epochs)
ax.tick_params(axis='both', labelsize=14)
ax.grid(True, axis="y", linestyle="--", alpha=0.3)
ax.legend(fontsize=13, ncol=2,  columnspacing=1, handletextpad=0.4, framealpha=0.94, loc="lower left")
plt.tight_layout()
plt.show()
```
